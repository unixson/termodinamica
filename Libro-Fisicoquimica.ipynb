### Ecuaciones de estado.

**La ley de Boyle**. Roberto Boyle encontró que para una cierta cantidad de gas a temperatura constante el volumen ocupado por el gas diminuye conforme aumenta la presión, en otras palabras, es inversamente proporcional a la presión, en términos matemáticos, $ V\; \alpha\; \frac{1}{p} $, para quitar el símbolo de proporcionalidad se introduce una constante, k,
$$
V = k\left(\frac{1}{P}\right)
$$

Esta ecuación es válida a temperatura constante (proceso isotérmico) y masa fija de gas.

La grafica de V vs. P, llamada isoterma, es construye facilmente empleando el siguiente codigo en python:

```python
import numpy as np
import matplotlib.pyplot as plt

# considerando que a 273.15 K  el volumen ocupado a 1 atm de presión es 22.4 L
p = np.arange(0.8,5,0.2)
V = 22.4/p
V1 = 40.51/p
V2 = 60.5/p
plt.plot(p,V,'-p')
plt.plot(p,V1, '-ok')
plt.plot(p,V2,'-v')
plt.xlabel("Presión /atm")
plt.ylabel("Volumen /litros")
plt.show()
```

**La ecuación de Charles-Gay-Lussac** por su parte estos investigadores estudiaron la relación del volumen con la temperatura a presión constante. Encontrando que este es directamente proporcional a la temperatura, es decir, el volumen aumenta conforme la temperatura aumenta. Por lo tanto, es una relación lineal entre la temperatura y el volumen. En términos matemáticos, la ecuación que relaciona a estas variables tiene la forma:
$$
V = a + bt
$$
donde a es el volumen a t=0 ($V_0$), y b es la pendiente de la linea recta.
$$
m = \frac{\Delta V}{\Delta t} = \left(\frac{\delta V}{\delta t}\right)_p
$$

Pero aún mas importante es el hallazgo el aumento relativo del volumen por cada grado centígrado es el mismo para todos los gases, el cual considerado con respecto al volumen $V_0$ es el coeficiente de expansión térmica ($\alpha_0$). 
$$
\alpha_0 = \frac{1}{V_0}\left(\frac{\delta V}{\delta t}\right)_p
$$
Resolviendo para la pendiente m:
$$
m=\left(\frac{\delta V}{\delta t}\right)_p= \alpha_0 V_0
$$
$$
V_0 = V_0 + V_0\alpha_0t = V_0\alpha_0 \left(\frac{1}{\alpha_0}+t \right)
$$
Si graficamos esta última ecuación con el siguiente codigo en Python

```python
import numpy as np
import matplotlib.pyplot as plt
vo=22.4
v1 = 24.46
v2 = 26.51
a = 1/273.15
t = np.arange(-274,40,15)
v = vo + vo*a*t
va=v1+v1*a*t
vb = v2+v2*a*t
plt.plot(t,v,'-p', label='273.15 K')
plt.plot(t,va, '-ok', label='298.15 K')
plt.plot(t,vb, '-o', label= '323.15 K')
plt.xlabel("temperatura/$^{\circ}C$")
plt.ylabel("Volumen")
plt.grid()
plt.show()
```

Considerando la gráfica anterior notamos que el volumen tiende a cero conforme nos acercamos a $-273.15\;^{\circ}C$. Entonces, 

$$
V = V_0 + V_0\alpha_0t = V_0\alpha_0 \left(273.15 + t \right)
$$
la relación $t+273.15$ es una nueva escala de temperatura, llamada del gas ideal donde $T/K=t/^{\circ}C+273.15$ y corresponde con la escala absoluta que actualmente se llama temperatura termodinámica y se reporta en Kelvin. Por lo tanto la ecuación empleando la temperatura del gas ideal o absoluta es,

$$
V = V_0\alpha_0T
$$


#### La ecuación general del estado gaseoso

Podemos construir una relación general combinando las ecuaciones de Boyle y Charles, considerando que $V_0$ es el volumen a cero grados centigrados. La ley de Boyle para esas condiciones es:

$$
V_0 = \frac{C_0}{p}
$$

sustitutendo en la ecuación $V = V_{0} \alpha_0T$ tenemos

$$
V= \frac{C_{0}\alpha_{0} T}{p}
$$

Ahora notamos que cuando la masa de gas aumenta manteniendo la temperatura y presión constante, el volumen aumenta proporcionalmente, por lo tanto la constante debe ser proporcional a la masa del gas(w) $C_0 = B w$. sustituyendo en la ecuacion anterior,

$$
V= \frac{B \alpha_{0} w T}{p}
$$

Esta es una ecuación que involucra las variables que definen el estado de un gas, sin embargo, no es general pues se requiere conocer el valor de la constante B, la cual es una propiedad particular para cada gas y se requeriria una tabla de valores para poder emplearla. Esto se puede remediar de la siguiente manera:

Primero despejamos w de la ecuación 

$$
w = \frac{1}{\alpha_0 B}\left(\frac{P V}{T}\right)
$$

y definimos todas las variables en un estado particular, por conveniencia puede ser P= 1 atm, V= 22,4 L, T= 273.15 K. Condiciones que llamaremos estandar.

$$
w = \frac{1}{\alpha_0 B}\left(\frac{P_0 V_0}{T_0}\right)
$$

Dado que las condiciones tipo son fijas, estas podemos definir una constante (la llamaremos con la letra R):

$$
R = \left(\frac{P_0 V_0}{T_0}\right)
$$

Entonces 

$$
w = \frac{R}{\alpha_0 B}
$$

Cuando tenemos fijas las condiciones de referencia, entonces tendremos una masa característica de cada gas, M, es decir la masa de 22.4 L de gas a p= 1 atm y T = 273.15K:

$$
M = \frac{R}{\alpha_0 B}
$$

Despejando B:
$$
B = \frac{R}{\alpha_0 M}
$$

sustituyendo en la ecuación

$$
V= \frac{B \alpha_{0} w T}{p}
$$

tendremos 
$$
V = \left(\frac{w}{M}\right)\frac{RT}{p}
$$

La masa característica, M, es el peso molecular del gas  $n=\frac{w}{M}$, por lo cual la ecuación queda:

$$
pV = nRT
$$


```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def g(w,M,T):
    plt.figure(2)
    p = np.arange(0.1,2.0, 0.05)
    plt.plot(p,w*0.08205*T/(p*M),'-o')
    plt.xlabel('presion/atm')
    plt.ylabel('Volumen/L')
    plt.ylim(0,50)
    plt.grid()
    plt.show

interactive_grafica=interactive(g, w=(1,40), M=(1,100),T=(100,400))
output = interactive_grafica.children[-1]
output.layout.height = '350px'
interactive_grafica
```

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#entering data as dictionary then convert to pandas data frame, but they were corrected to show some dependence of pressure.
densidad ={'density':[20.847,72.54,134.23,205.92,300.62],'pressure':[1E3,3E3,5E3,7E3,9E3]}
densidad_data = pd.DataFrame(data=densidad)
densidad_data
```

```python
#plot data
x=densidad_data.pressure
y = (densidad_data.density)/densidad_data.pressure
plt.scatter(x,y)
```

```python
# modelar o ajustar los datos
model =np.polyfit(x,y,1)
model
```

```python
# predecir los datos con el modelo
predict = np.poly1d(model)
presion = 1E4
predict(presion)
```

```python
#ahora predecimos los datos para un modelo
x_line_range = np.arange(0, 1E4,1000)
y_line_range = predict(x_line_range)
#graficamos
plt.scatter(x,y)
plt.plot(x_line_range,y_line_range)
plt.ylim(0.015,0.035)
plt.ylabel('$\delta$/p')
plt.xlabel('p/atm')
print('El intercepto es',model[1], 'con una pendiente',model[0], 'el peso molecular es',model[1]*8.314*294.25)
```
